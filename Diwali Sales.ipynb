```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Set matplotlib to display plots inline in the notebook
%matplotlib inline
```

```python
# Define the path to the CSV file
path = r"C:\Users\kimotho\Downloads\Projects\Diwali\Diwali Sales Data.csv"

#Read the CSV fole into a DataFrame called 'sales'
sales = pd.read_csv(path,encoding= 'unicode_escape' )
sales
```

```python
# Display information about the DataFrame
sales.info()
```

```python
# Display the number of rows and columns of the DataFrame
sales.shape
```

```python
# Drop unecessary columns
sales = sales.drop(columns =['unnamed1', 'Status'])
sales
```

```python
# Display information about the DataFrame
sales.info()
```

```python
# Convert the 'User_ID' column to object(string) data type
sales['User_ID']=sales['User_ID'].astype(object)
```

```python
sales.info()
```

```python
# Display descriptive statistics of the DataFrame
sales.describe()
```

## Univariate

```python
# Create a count plot for the 'Gender' column in the 'sales' DataFrame
ax= sb.countplot(data=sales, x='Gender')

# Add labels to the bars in the count plot
for bars in ax.containers:
    ax.bar_label(bars);
```

```python
# Set the color palette for the count plot
color =sb.color_palette()[0]

# Create a count plot for the 'Occupation' column in the 'sales' DataFrame
# The order of categories is based on the value counts of 'Occupation'
order = sales['Occupation'].value_counts().index
sb.countplot(data=sales, y='Occupation', order=order, color=color)

# Rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

Most buyers are from the IT Sector, Healthcare and aviation respectively

## Bivariate

```python
# Create a count plot for the 'Age Group' column in the 'sales' DataFrame,
# with bars colored by 'Gender'
ax= sb.countplot(data=sales, x='Age Group', hue='Gender')

# Add labels to the bars in the count plot
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Group the 'sales' DataFrame by 'Age Group' and calculate the sum of 'Amount'
# Sort the groups by the sum of 'Amount' in descending order
cx = sales.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax= sb.barplot(data=cx, x='Age Group', y='Amount')
```

```python
# Group the 'sales' DataFrame by 'State' and calculate the sum of 'Orders'
# Sort the groups by the sum of 'Orders' in descending order
dx=sales.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
color=sb.color_palette()[0]
ax= sb.barplot(data=dx, x='State', y='Orders', color=color)

# Add labels to the bars in the count plot
for bars in ax.containers:
    ax.bar_label(bars)

# Rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90)

# Set the figure size for the plot
rcParams['figure.figsize']=14, 6
```

- Most of the orders were from Uttar Predesh, Maharashtra and Karnataka respectively

```python
# Create a count plot for the 'Marital_Status' column in the 'sales' DataFrame,
ex=sb.countplot(data=sales, x='Marital_Status',color=color)

# Add labels to the bars in the count plot
for bars in ex.containers:
    ex.bar_label(bars)
```

```python
# Group the 'sales' DataFrame by 'Marital_Status' and 'Gender' columns
# Calculate the sum of 'Amount' for each group
# Sort the groups by the sum of 'Amount' in descending order

mar=sales.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create a bar plot for the grouped data, showing the sum of 'Amount' for each 'Marital_Status' and 'Gender'
sb.barplot(data=mar, x='Marital_Status', y='Amount', hue='Gender');
```

- from the above graph, we can see that most of the buyers are unmaried women

### Exploring Product Category

```python
# Group the 'sales' DataFrame by 'Product_Category' and calculate the sum of 'Amount'
# Sort the groups by the count of occurrences in descending order
order=sales['Product_Category'].value_counts().index
cx=sb.countplot(data=sales, x='Product_Category', order=order, color=color)

# Add labels to the bars in the count plot
for bars in cx.containers:
    cx.bar_label(bars)

# Rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90);
```

```python
# Group the 'sales' DataFrame by 'Product_Category' and calculate the sum of 'Amount'
# Sort the groups by the sum of 'Amount' in descending order
px=sales.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

# Create a bar plot for the grouped data, showing the sum of 'Amount' for each 'Product_Category'
sb.barplot(data=px, x='Product_Category', y='Amount', color=color)

# Rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90);
```

- The top five selling products are food, clothing&apparel, electronics& gadgets, footware&shoes and furniture
